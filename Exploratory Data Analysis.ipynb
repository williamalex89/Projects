### INTRODUCTION:

Strong student habits, particularly those related to studying and time management, are strongly correlated with improved academic performance. Students who actively engage in learning, manage their time effectively, and prioritize their well-being are more likely to achieve higher grades and academic success.

### ABOUT THE DATASET:
This is a simulated dataset exploring how lifestyle habits affect academic performance in students. Based on this student records and 10+ features including study hours, sleep patterns, social media usage, diet quality, mental health, and final exam scores, we will analyse how each factor affect the performance of students.
This dataset simulates 1,000 students' daily habits—from study time to mental health—and compares them to final exam scores. It's like spying on your GPA through the lens of lifestyle.

### Age wise exam Score of Male & Female students using Pivot. Age range falls between 17 to 24.

```python
data.pivot_table(values = 'exam_score', index = 'age', columns = 'gender', aggfunc = 'mean',margins = True).round(2) ## Need 2 Categorical Columns & 1 Numerical column
```

##### Here we find that , the age group of 18 has got comparatively high scores in exam which is 70.93 %. Students from age 21 has scored less in exams where the score is 67.54%.Moreover, the average score of sudents in this data set is 69.60%.

### Analysing age wise mental health and academic performance.

##### Students with better mental health (10) has got a better score in exam of 77.95% comparing to the ones with least mental health (1) having 62.37%. So, mental health is an imprortant factor that affects the academic performance.
